```python
import pandas as pd
```

# Load Data

```python
data_df = pd.read_csv('..\Data\Raw\VAERSDATA.csv')
vax_df = pd.read_csv('..\Data\Raw\VAERSVAX.csv')
symptoms_df = pd.read_csv('..\Data\Raw\VAERSSYMPTOMS.csv')
```

```python
print("Shapes:")
print("Data:", data_df.shape)
print("Vax:", vax_df.shape)
print("Symptoms:", symptoms_df.shape)
```

```python
print("Columns:")
print("Data:", data_df.columns) 
print("Vax:", vax_df.columns)
print("Symptoms:", symptoms_df.columns)
```

# Merge Dataframes

```python
merged_data_vaxed_df = pd.merge(data_df, vax_df, on='VAERS_ID', how='left')
merged_data_vaxed_df
```

```python
merged_data_vaxed_symp_df = pd.merge(merged_data_vaxed_df, symptoms_df, on='VAERS_ID', how='left')
merged_data_vaxed_symp_df
```

# Save Dataframe

```python
merged_data_vaxed_symp_df.to_csv('..\Data\Processed\merged_data_vaxed_symp.csv', index=False)
```
