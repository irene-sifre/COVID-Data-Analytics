# Cleaning Vaccines CSV

```python
import pandas as pd
vax_df = pd.read_csv('../Data/VAERSVAX.csv', low_memory=False)
```

```python
vax_df.head(20)
```

```python
vax_df.columns
```

```python
drop_columns = ["VAX_ROUTE","VAX_SITE","VAX_TYPE","VAX_NAME"]
vax_df = vax_df.drop(columns=drop_columns)
```

```python
vax_df.head()
```

```python
vax_df['VAX_MANU'].value_counts()
```

```python
# Remove rows where 'VAX_MANU' is 'UNKNOWN MANUFACTURER'
vax_df = vax_df.loc[vax_df['VAX_MANU'] != 'UNKNOWN MANUFACTURER']
```

```python
vax_df.head(20)

```

```python
vax_df.info()
```

```python
vax_df = vax_df.dropna()
```

```python
vax_df.head(20)
```

```python
# Remove rows where 'VAX_DOSE_SERIES' is 'UNK'
vax_df = vax_df[vax_df['VAX_DOSE_SERIES'] != 'UNK']

# Display the resulting DataFrame to verify
vax_df
```

```python
# Convert all values to uppercase and remove spaces in 'VAX_LOT'
vax_df['VAX_LOT'] = vax_df['VAX_LOT'].str.upper().str.replace(' ', '')

# Display the resulting DataFrame to verify
vax_df
```

```python
# Get unique values in the VAX_LOT' column
unique_values_vax_lot = vax_df['VAX_LOT'].unique()

# Display the unique values
unique_values_vax_lot
```

```python
# Get unique values along with their frequencies in the 'VAX_LOT' column
unique_values_with_counts = vax_df['VAX_LOT'].value_counts()

# Display the unique values along with their frequencies
print(unique_values_with_counts)
```

```python
vax_df = vax_df[~vax_df['VAX_LOT'].str.contains(r'[^a-zA-Z0-9\s]', regex=True, na=False)]
```

```python
# Get unique values along with their frequencies in the 'VAX_LOT' column
unique_values_with_counts = vax_df['VAX_LOT'].value_counts()

# Display the unique values along with their frequencies
print(unique_values_with_counts)
```

```python
# Remove rows where VAX_LOT is 'UNKNOWN'
vax_df = vax_df[vax_df['VAX_LOT'] != 'UNKNOWN']

# Display the cleaned dataframe
print(vax_df['VAX_LOT'].value_counts())
```

```python
# Get counts of each unique value in 'VAX_LOT'
lot_counts = vax_df['VAX_LOT'].value_counts()

# Filter to get only the lots with more than 3000 entries
lots_ = lot_counts[lot_counts > 3000].index
```

```python
import matplotlib.pyplot as plt

# Filter to get only the lots with more than 3000 entries
lots_more_than_3000 = lot_counts[lot_counts > 3000]

# Plot the data
plt.figure(figsize=(15, 5))
lots_more_than_3000.plot(kind='bar')
plt.title('Vaccine Lots with More Than 3000 Entries')
plt.xlabel('Vaccine Lot')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# Get counts of each unique value in 'VAX_LOT'
lot_counts = vax_df['VAX_LOT'].value_counts()

# Filter to get only the lots with more than 3000 entries
lots_more_than_3000 = lot_counts[lot_counts > 3000].index
```

```python
# Filter the DataFrame to keep only rows where 'VAX_LOT' is in the list of lots with more than 3000 entries
vax_df = vax_df[vax_df['VAX_LOT'].isin(lots_more_than_3000)]

# Display the resulting DataFrame to verify
vax_df
```

```python
# Save the cleaned dataframe to a CSV file
output_file_path = '../Data/Clean/vax_data.csv'
vax_df.to_csv(output_file_path, index=False)
```
